```python
y=train["target"]
X=train.drop("target",axis=1)
X
```

```python
X=X.drop("era",axis=1)
X=X.drop("data_type",axis=1)
X
```

```python
good_cols=[col for col in X.columns if not col.startswith("target")]
 
X=X[good_cols]
X
```

```python
X=X.reset_index(drop=True)
y=y.reset_index(drop=True)
```

```python
X.isnull().values.any()
```

```python
def kinetic(x):
    probs=np.unique(x,return_counts=True)[1]/len(x)
    kinetic=np.sum(probs**2)
    return  kinetic
```

```python
X["KINETIC_FEAT"]=X.apply(kinetic,axis=1)

X["max_FEAT"]=X[good_cols].apply(np.max,axis=1)

X["mean_FEAT"]=X[good_cols].apply(np.mean,axis=1)
```

```python
del train
gc.collect()
```

```python
X["KINETIC_FEAT"]=np.round(X["KINETIC_FEAT"], decimals=2)
X["mean_FEAT"]=np.round(X["mean_FEAT"], decimals=2)
```

```python
X["KINETIC_FEAT"]=( X["KINETIC_FEAT"]% 1 * 100).astype(int)
X["mean_FEAT"]=( X["mean_FEAT"]% 1 * 100).astype(int)
```

```python
model = lgb.LGBMRegressor(
    n_estimators=50,
    learning_rate=0.01,
    max_depth=5,
    num_leaves=2**5 - 1,
    colsample_bytree=0.8,  
    subsample=0.8,
    random_state=42
)
```
