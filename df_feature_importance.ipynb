```python
X.isnull().values.any()
```

```python
def  kin_energy(x):
    f=np.unique(x,return_counts=True)
    prob=f[1]/x.shape[0]
    energy=np.sum(prob**2)
    return  energy
```

```python
def count_common(x):
    a=np.array(x)
    b=np.where(a==np.min(a))
    v=np.array(b)
    nr_ap=v.shape[1]
    return nr_ap
```

```python
def count_least(x):
    a=np.array(x)
    b=np.where(a==np.max(a))
    v=np.array(b)
    nr_ap=v.shape[1]
    return nr_ap
```

```python
X["KINETIC_FEAT"]=X.apply(kin_energy,axis=1)

X["max_FEAT"]=X[good_cols].apply(np.max,axis=1)

X["mean_FEAT"]=X[good_cols].apply(np.mean,axis=1)
X["COMMON_COUNT_FEAT"]=X.apply(count_common,axis=1)
X["LEAST_COUNT_FEAT"]=X.apply(count_least,axis=1)
```

```python
X["KINETIC_FEAT"]=np.round(X["KINETIC_FEAT"], decimals=2)
X["mean_FEAT"]=np.round(X["mean_FEAT"], decimals=2)
```

```python
X["KINETIC_FEAT"]=( X["KINETIC_FEAT"]% 1 * 100).astype(int)
X["mean_FEAT"]=( X["mean_FEAT"]% 1 * 100).astype(int)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42)
```

```python
model = lgb.LGBMRegressor(
    n_estimators=50,
    learning_rate=0.01,
    max_depth=5,
    num_leaves=2**5 - 1,
    colsample_bytree=0.8,  
    subsample=0.8,
    random_state=42
)
```

```python
model.fit(X_train, y_train)
```
