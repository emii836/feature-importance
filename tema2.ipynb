```python
def produs(x):
    elm=np.array(x)
    e=np.sort(elm)[::-1]
    produs=e[0]*e[1]
    return produs
```

```python
 def produs_min(x):
     elm=np.array(x)
     ar=np.sort(elm)
     produs=ar[0]*ar[1]
     return produs
```

```python
X["PROD_MAX_FEAT"]=X.apply(produs,axis=1)
X["PROD_MIN_FEAT"]=X.apply(produs_min,axis=1)
```
