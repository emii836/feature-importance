```python
def kinetic(x):
    probs=np.unique(x,return_counts=True)[1]/len(x)
    kinetic=np.sum(probs**2)
    return  kinetic
```

```python
def most(x):
    uniques, counts=np.unique(x,return_counts=True)
    return uniques[np.argmax(counts)]
```

```python
def count_elem(u):
    uniques, counts=np.unique(u,return_counts=True)
    return counts[np.argmin(uniques)]
    
```

```python
def least(x):
    uniques, counts=np.unique(x,return_counts=True)
    return uniques[np.argmin(counts)]
```

```python
def count_list(u):
    uniques, counts=np.unique(u,return_counts=True)
    return counts[np.argmax(uniques)]
    
```

```python
X["LEAST_FEAT"]=X.apply(least,axis=1)
X["COUNTLEAST_FEAT"]=X.apply(count_list,axis=1)
X["KINETIC_FEAT"]=X.apply(kinetic,axis=1)
X["max_FEAT"]=X[good_cols].apply(np.max,axis=1)
X["mean_FEAT"]=X[good_cols].apply(np.mean,axis=1)
X["MOST_FEAT"]=X.apply(most,axis=1)
X["COUNT_FEAT"]=X.apply(count_elem,axis=1)
```
